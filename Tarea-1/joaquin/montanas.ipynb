




























## Carga

```python
import pandas as pd
import plotly.express as px

# Carga
df = pd.read_csv('data.csv')
```

## Limpieza y Agrupación

```python
# Limpieza y variables
cols = ['Age', 'Anxiety', 'While working']
df_clean = df[cols].dropna().copy()

# Grupos etarios
bins = [0, 18, 25, 40, 100]
labels = ['Adolescentes (0-18)', 'Jóvenes (19-25)', 'Adultos (26-40)', 'Mayores (41+)']
df_clean['Etapa de Vida'] = pd.cut(df_clean['Age'], bins=bins, labels=labels)
```

## Gráfico

```python
# Gráfico
fig = px.violin(
    df_clean,
    x='Anxiety',
    y='Etapa de Vida',
    color='While working', 
    orientation='h',
    points=False,
    title='¿Influye la música al trabajar según la edad?',
    category_orders={
        'Etapa de Vida': ['Mayores (41+)', 'Adultos (26-40)', 'Jóvenes (19-25)', 'Adolescentes (0-18)']
    },
    color_discrete_map={
        'Yes': '#00B4D8',
        'No': '#7209B7'
    }
)

# Elevación
fig.update_traces(
    side='positive',   
    width=2.2,         
    meanline_visible=True,
    opacity=0.8        
)

# Estilo
fig.update_layout(
    xaxis_title='Nivel de Ansiedad (0 - 10)',
    yaxis_title='',
    legend_title='¿Escucha música al trabajar?',
    plot_bgcolor='white',
    height=650,
    font=dict(size=14)
)

fig.show()
```
